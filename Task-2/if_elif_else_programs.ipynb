**# Program 1: Check whether a number is positive or negative**

```python
num = float(input("Enter the number: "))
if(num>0):
  print("Your number is positive")
elif(num<0):
  print("Your number is negative")
else:
  print("Your number is neither positive nor negative its Zero")
```

**# Program 2: Check Voting Eligibility(Check whether a person is eligible to vote)**

```python
age = int(input("Enter your age: "))
if(age>=18):
  print("You are eligible for voting")
else:
  print("You are not eligible for voting because you are kid")
```

**# Program 3: Check Whether a Number is Even or Odd**

```python
num = int(input("Enter the number: "))
if(num%2==0):
  print("Your number is even")
else:
  print("Your number is odd")
```
