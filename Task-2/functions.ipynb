**Program 1: Function with Parameters**

```python
def greet(name):
    print("Hello", name)

greet("Sajal")
```

**Program 2: Function with Return Value**

```python
def add(a, b):
    return a + b

result = add(25, 28)

print("Sum =", result)
```

**Program 3: Default Parameters**

```python
def welcome(name="Sir"):
    print("Welcome", name)

welcome()

welcome("Sajal")
```

**Program 4: Local and Global Variables**

```python
x = 100

def demo():
    y = 50

    print("Global Variable:", x)
    print("Local Variable:", y)

demo()
```
