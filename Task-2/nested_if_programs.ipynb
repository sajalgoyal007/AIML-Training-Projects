**# Program 1: College Admission Eligibility**

```python
marks = int(input("Enter your marks: "))
age = int(input("Enter your age: "))

if(marks>=60):

  if(age>=17):
    print("You are eligible for college admission")
  else:
    print("You are not eligible for college admission because you are below 17")

else:
  print("You are not eligible because marks are below 60")
```

**# Program 2: ATM Withdrawal Validation**

```python
balance = 10000
pin = 1234
entered_pin = int(input("Enter your pin: "))
amount = float(input("Enter the withdrawal amount: "))

if entered_pin == pin:
  if(amount<=balance):
    print("Withdrawal successful")
    print("Your remaining balance is ", balance-amount)
  else:
    print("Insufficient balance")

else:
  print("Incorret pin")
```

**# Program 3: Employee Bonus Eligibility**

```python
salary = float(input("Enter the salary:"))
rating = input("Enter performance rating (good/average/poor): ").lower()

if salary >= 25000:
  if rating == 'good':
    print("Employee is eligible for a 10% bonus")
  elif rating == 'average':
    print("Employee is eligible for 5% bonus")
  else:
    print("Employee is not eligible for bonus due to its work (rating performance)")

else:
  print("Employee is not eligible for bonus")
```
