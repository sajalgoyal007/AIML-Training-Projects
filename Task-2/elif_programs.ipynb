**# Program 1: Student Grade Classification**

```python
marks = int(input("Enter you marks"))
if marks >= 90:
    print("Grade A+")

elif marks >= 75:
    print("Grade A")

elif marks >= 60:
    print("Grade B")

elif marks >= 40:
    print("Grade C")

else:
    print("Grade D ---> Fail")
```

**# Program 2: Traffic Signal Simulation**

```python
signal = input("Enter traffic signal color: ").lower()

if signal == "red":
  print("Stop")
elif signal == "yellow":
  print("Get ready")
elif signal == "green":
  print("Go")

else:
  print("Invalid signal")
```

**# Program 3: Weather Recommendation System**

```python
weather = input("Enter weather condition (sunny/rainy/cold): ").lower()

if weather == "sunny":
  print("You can go for the walk with me")
elif weather == "rainy":
  print("You can go for the walk with me but with take an umbrella")
elif weather=="cold":
  print("You should Wear warm clothes")
else:
  print("Invalid weather condition")
```
