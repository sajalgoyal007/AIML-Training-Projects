**Program 1: Login Validation**

```python
username = input("Enter username: ")
password = input("Enter password: ")
if username == "sajal" and password == "2007":
  print("Login Successfully")
elif username == "sajal" and password != "2007":
  print("Incorrect Password")
elif username != "sajal" and password == "2007":
  print("Incorrect Username")
else:
  print("Incorrect Username and Password")
```

**Program 2: Loan Eligibility Checker**

```python
age = int(input("Enter your age: "))
salary = float(input("Enter your salary: "))
credit_score = int(input("Enter your credit score: "))

if age<21:
  print("Rejected! because age should be at least 21")
elif salary<25000:
  print("Rejected! salary is too low")
elif credit_score<600:
  print("Rejected! poor credit score")

else:
  print("Loan Approved!!")
```

**Program 3: Scholarship Eligibility System**

```python
marks = int(input("Enter your marks(0-100): "))
income = float(input("Enter your family income(per annum): "))

if marks>90:
  print("Not eligible! because marks are below 90")
elif income>300000:
  print("Not eligible! because family income is higher than the criteria")

else:
  print("Eligible for Scholarship!!")
```
