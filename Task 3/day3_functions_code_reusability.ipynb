### Topic 6:-
* Variable Scope (Local, Global, Nonlocal)
    * Variable Scope --> where a specific variable can be seen and accessed within your code/program
    * Python use different scope to to protect data from being accidently changed and mixed by diffenents parts of the programs
    

    * Local Scope:
        * Variables created inside a function
        * Accessible only inside function
        *  They only exist while that function runs and disappear once it finishes
    * Global Scope:
        * Variables created outside all functions
        * They are accessible by any function anywhere in the file
    * NonLocal Scope:
        * Variables used specifically in nested (inner) functions
        * It connects to variable of outer function
        * The nonlocal keyword is required because without it, the inner function would create a new local variable instead of modifying the outer one
        * but not for the global variable
        * Example:  
```python
def outer():

    x = 10

    def inner():

        nonlocal x
        x = 20

    inner()

    print(x)

outer()
```

**1.  BMI Calculator Function**
* Create a reusable function with parameters for height and weight.
* Display BMI value and category.

````python
def bmi_calculator(weight, height):

    bmi = weight / (height ** 2)

    print("BMI =", round(bmi, 2))

    if bmi < 18.5:
        print("Category: Underweight")

    elif bmi < 25:
        print("Category: Normal Weight")

    elif bmi < 30:
        print("Category: Overweight")

    else:
        print("Category: Obese")


weight = float(input("Enter your weight (kg): "))
height = float(input("Enter your height (m): "))

bmi_calculator(weight, height)
````

**2. EMI Calculator**
* Calculate monthly EMI based on loan amount, interest rate, and tenure.

````python
def emi_calculator(loan, rate, months):

    interest = (loan * rate * months) / 100
    total = loan + interest
    emi = total / months

    print("Monthly EMI =", round(emi, 2))


loan = float(input("Enter Loan Amount: "))
rate = float(input("Enter Interest Rate: "))
months = int(input("Enter Number of Months: "))

emi_calculator(loan, rate, months)
````

**3. Recursive Factorial Program**
* Implement factorial calculation using recursion.

````python
def factorial(n):

    if n == 1 or n == 0:
        return 1

    return n * factorial(n - 1)


num = int(input("Enter a number: "))

print("Factorial =", factorial(num))
````

**4. Fibonacci Series Generator**

* Generate Fibonacci numbers using functions.
* Bonus: Implement using recursion.

````python
def fibonacci(n):

    a = 0
    b = 1

    for i in range(n):
        print(a, end=" ")

        c = a + b
        a = b
        b = c


num = int(input("Enter number of terms: "))

fibonacci(num)
````

**5. Prime number checker**

* Create a function to determine whether a number is prime.

````python
def prime_checker(num):

    if num <= 1:
        print("Not a Prime Number")
        return

    for i in range(2, num):

        if num % i == 0:
            print("Not a Prime Number")
            return

    print("Prime Number")


number = int(input("Enter a number: "))

prime_checker(number)
````

**6. List operation functions**   
Demonstrate the use of:    
map()     
 filter()     
reduce()

````python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# map()
square = list(map(lambda x: x * x, numbers))
print("Map:", square)

# filter()
even = list(filter(lambda x: x % 2 == 0, numbers))
print("Filter:", even)

# reduce()
total = reduce(lambda x, y: x + y, numbers)
print("Reduce:", total)
````
