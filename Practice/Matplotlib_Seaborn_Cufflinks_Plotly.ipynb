Object Oriented way

Matplolib allow the aspect ratio, DPI and figure size to be specified when the figure object is created. You can use the figsize and dpi keyword argument       
figsize is the tuple of the width and height of the figure in inches dpi is the dop-per-inch(pixel-per-inch)

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1, 0.1, 1, 1])   # gap from left, gap from right, 80% will be the width, 80% will be the height
axis1.plot(x, y)
axis2 = fig.add_axes([0.3, 0.4, 0.3, 0.3])
axis2.plot(y, x)
```

```python
fig = plt.figure(figsize=(6, 8), dpi=120)
axis1 = fig.add_axes([0.1, 0.1, 1, 1])
axis1.plot(x, y)
axis2 = fig.add_axes([0.3, 0.4, 0.3, 0.3])
axis2.plot(y, x)
```
