## 1. Classes and Objects

### What is a Class?

A class can be thought of as a blueprint, map, or recipe for creating objects. It does not occupy any physical space or memory on its own. Instead, it defines the attributes (data) and methods (functions) that objects created from it will have.

In simple words, a class describes what an object should look like and what actions it can perform.

### Real-Life Example

Suppose there is a Student Form in a college.

The form contains fields like:    

Name      
Roll Number   
Branch  

The form itself is only a structure that defines what information should be stored.

Similarly, a class defines what data and functions an object will have.

---

### What is an Object?

An object is a real instance created from a class. It stores actual values and occupies memory.

### Real-Life Example

If Student is a class, then:

Sajal   
Jogendra   
Priya   

are different objects of the Student class.

Each student has different data but follows the same structure defined by the class.

---

### Difference Between Class and Object

| Class | Object |
|---------|---------|
| A blueprint or template           | An instance of a class |
| Defines structure and behavior    | Contains actual data   |
| Created using the `class` keyword | Created from a class   |
| Does not store individual data    | Stores individual data |

## 6. Inheritance
* It is an main OOPs Concept in which one class receive the properties and method of another class
* The class whose properties is inherited is called the parent class(base class) and the class which inherit those properties is calles child class(derived class)
* Inheritance help in code reusability because if we use this then we do not have to write the same code again and again
* We use inheritance because to use existing the code, reduce code duplication, it make programs easier to maintain and to create the hierarchical relationship between the classes
* Example:
```python
class Person:
    def show(self):
        print("I am a Person")
class Student(Person):
    pass
s1 = Student()
s1.show()

its output is-> I am a Person

here person is the parent class and student is the child class and Student inherits from the Person, in this Student doesn't have its own show() method, Still, it can use show() because it is inherited from the parent class.
```
* Types of inheritance
    1.   Single Inheritance:
    * In this having only one child class and one parent class and the one child class is inherits from the one parent class
    ```python
  class A:
     pass
  class B(A):
      pass
    ```
    2.   Multiple Inheritance:
    * A class inherits from a class that already inherits from another class
    ```python
    class A:
      pass
    class B(A):
      pass
    class C(B):
      pass
    ```














    3. Multiple Inheritance:
    * A child class inherits from more than one parent class

    ```python
        pass
    ```

##8. Abstraction
* Abstraction is the process of hiding unnecessary implementation details and showing only the essential features / details to the user
* abstraction focuses on **what an object does rather than how it does it**
* We use abstraction because to reduce the complexity, reduce code redability, hide implementation details and focus only essential features
1. Abstract class:
    * It is the class that contain one or more abstract method
    * A abstract method can not be initialized directly
    * This class created using the ABC module
2. Abstract method:
    * An abstract method is a method that is declared but does not contain any implementation
    * The child class must provide the implementation of the abstract method
    * Abstract method is created using the @abstractmethod

* Example:
```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Circle(Shape):
    def area(self):
        print("Area of Circle")
c = Circle()
c.area()

its output is
Area of Circle

Here Shape is the abstract class and area() is the abstract method and Circle inherits from Shape and Circle() provide implementation of area() and Therefore, an object of Circle can be created.
```

````python
class Student:

    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def calculate_grade(self):

        if self.marks >= 90:
            return "A"

        elif self.marks >= 75:
            return "B"

        elif self.marks >= 60:
            return "C"

        else:
            return "D"

    def display_details(self):

        print("Name:", self.name)
        print("Roll No:", self.roll_no)
        print("Marks:", self.marks)
        print("Grade:", self.calculate_grade())


student1 = Student("Sajal", 101, 88)

student1.display_details()
````

**Assignment 2: Bank Account System (Encapsulation)**

````python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance
    def deposit(self, amount):
        self.__balance += amount
        print(amount, "deposited successfully")
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(amount, "withdrawn successfully")
        else:
            print("Insufficient Balance")
    def get_balance(self):
        return self.__balance

account = BankAccount(5000)
account.deposit(2000)
account.withdraw(1000)

print("Current Balance:", account.get_balance())
````

**Assignment 5: Shape Area Calculator (Abstraction)**

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        return self.length * self.breadth

rectangle = Rectangle(10, 5)

print("Area:", rectangle.area())
````

**Assignment 6: Library Management System (Mini Project)**

````python
class Book:

    def __init__(self, title):
        self.title = title
        self.issued = False

class Member:
    def __init__(self, name):
        self.name = name


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(book.title, "added successfully")

    def issue_book(self, title):
        for book in self.books:
            if book.title == title and not book.issued:
                book.issued = True
                print(title, "issued successfully")
                return
        print("Book not available")

    def return_book(self, title):
        for book in self.books:
            if book.title == title and book.issued:
                book.issued = False
                print(title, "returned successfully")
                return
        print("Book was not issued")

    def display_available_books(self):
        print("\nAvailable Books:")
        for book in self.books:
            if not book.issued:
                print(book.title)


library = Library()

book1 = Book("Python Programming")
book2 = Book("Machine Learning")

library.add_book(book1)
library.add_book(book2)

library.issue_book("Python Programming")

library.display_available_books()

library.return_book("Python Programming")

library.display_available_books()
````
